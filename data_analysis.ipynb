```python
import pandas as pd
import numpy as np
import torch
import torch_geometric

# Dodanie klasy do bezpiecznej listy (wymagane w PyTorch 2.6+)
torch.serialization.add_safe_globals([torch_geometric.data.data.Data])
```

```python
import torch

graphs = torch.load(
    "worldcup_2022_graphs.pt",
    map_location="cpu",
    weights_only=False
)
```

```python


pressure_values = [int(g.y_pressure.item()) for g in graphs]
turnover_values = [int(g.y_turnover.item()) for g in graphs]

print("Pressure distribution:")
print(pd.Series(pressure_values).value_counts())
print(pd.Series(pressure_values).value_counts(normalize=True))

print("\nTurnover distribution:")
print(pd.Series(turnover_values).value_counts())
print(pd.Series(turnover_values).value_counts(normalize=True))
```

```python
import numpy as np

def baseline_bce(y):
    y = np.array(y)
    p = y.mean()
    eps = 1e-8
    bce = -(y * np.log(p + eps) + (1 - y) * np.log(1 - p + eps)).mean()
    return p, bce

p_pressure, bce_pressure = baseline_bce(pressure_values)
p_turnover, bce_turnover = baseline_bce(turnover_values)

print("Pressure positive rate:", p_pressure)
print("Pressure baseline BCE:", bce_pressure)

print("Turnover positive rate:", p_turnover)
print("Turnover baseline BCE:", bce_turnover)
```
